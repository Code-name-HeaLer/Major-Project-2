```python
from __future__ import print_function, division
import time,os,sys,datetime

import IPython
import IPython.display

#Plotting tools
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
plt.style.use('ggplot')
rcParams['figure.figsize'] = (14, 14)
mpl.rcParams['axes.grid'] = False
%matplotlib inline

#Computation tools
import pandas as pd
import numpy as np
from six import iteritems
import tensorflow as tf


#Printing out the Versions of libraries 
print('Python Version: {}'.format(sys.version))
print('Tensorflow Version: {}'.format(tf.__version__))
print('Matplotlib Version: {}'.format(mpl.__version__))
print('Numpy Version: {}'.format(np.__version__))
print('Pandas Version: {}'.format(pd.__version__))
print('Seaborn Version: {}'.format(sns.__version__))
```

```python
#This function will load the activation list that was saved in .npz format

def load_npz_to_list(filename):
    loaded = np.load(filename, allow_pickle=True)
    return [loaded[key] for key in loaded]
```

# 1. Load the Healthy Activations Dataset

```python
REFIT_B2_DW_healthy_activations_list = load_npz_to_list("11_REFIT_B2_DW_healthy_activations.npz")

print("No. of healthy activations = ", len(REFIT_B2_DW_healthy_activations_list))
```

```python
column_names = np.array(["Aggregate Energy", "Synth Agg Energy", "Appliance Energy", "Appliance State"], dtype=object)
```

```python
#Plot an activation
act_no = 33

fig, axes = plt.subplots(len(column_names), 1, figsize=(6, 8), sharex=True)

# Plot each column as a separate subplot
for i in range(len(column_names)):
    axes[i].plot(REFIT_B2_DW_healthy_activations_list[act_no][:, i],label = column_names[i])
    axes[i].legend(loc="best")  # Add legend to each subplot
    
plt.xlabel("Time steps")  # Add x-axis label
plt.show()
```

## Comparison Plot

```python

# List of all lumpy arrays
arrays = [
    REFIT_B2_DW_healthy_activations_list[100][:, 2],
    REFIT_B2_DW_unhealthy_high_energy_activations[28][:, 2],
    REFIT_B2_DW_unhealthy_low_extended_energy_activations[28][:, 2],
    REFIT_B2_DW_unhealthy_repeated_cycle_activations[28][:, 2],
    REFIT_B2_DW_unhealthy_noisy_activations[28][:, 2],
    REFIT_B2_DW_unhealthy_noisy_activations[57][:, 2],
]

# Find the maximum length among all arrays
max_length = max(arr.shape[0] for arr in arrays)

# Function to pad each array with zeros
def pad_array(arr, target_length):
    return np.pad(arr, (0, target_length - arr.shape[0]), mode='constant', constant_values=0)

# Pad all arrays to the max length
padded_arrays = [pad_array(arr, max_length) for arr in arrays]

# Create DataFrame
df_for_plot = pd.DataFrame({
    'Healthy': padded_arrays[0],
    'Unhealthy with High Energy': padded_arrays[1],
    'Unhealthy with Extended low energy Cycle': padded_arrays[2],
    'Unhealthy with Repeated Cycles': padded_arrays[3],
    'Unhealthy with Noise 1': padded_arrays[4],  
    'Unhealthy with Noise 2': padded_arrays[5],  
})

print(df_for_plot.shape)  # Check the final shape
```

```python
# Define subplot layout (3 rows, 2 columns)
fig, axes = plt.subplots(3, 2, sharex=True, sharey=True, figsize=(18, 12))
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.weight"] = "bold"

df_1 = df_for_plot.reset_index()['Healthy']
df_2 = df_for_plot.reset_index()['Unhealthy with High Energy']
df_3 = df_for_plot.reset_index()['Unhealthy with Extended low energy Cycle']
df_4 = df_for_plot.reset_index()['Unhealthy with Repeated Cycles']
df_5 = df_for_plot.reset_index()['Unhealthy with Noise 1']
df_6 = df_for_plot.reset_index()['Unhealthy with Noise 2']

# Plot each DataFrame in a separate subplot
df_1.plot(ax=axes[0, 0], legend=False, color='green')
df_2.plot(ax=axes[0, 1], legend=False, color='red')
df_3.plot(ax=axes[1, 0], legend=False, color='purple')
df_4.plot(ax=axes[1, 1], legend=False,  color='blue')
df_5.plot(ax=axes[2, 0], legend=False, color='gray')
df_6.plot(ax=axes[2, 1], legend=False,  color='orange')

# Customize background color
for ax in axes.flat:
    ax.set_facecolor('white')
    plt.setp(ax.spines.values(), color='black')
#    ax.tick_params(axis='both', colors='black')
    ax.tick_params(axis='both', labelsize=25, colors='black')
    
axes[0, 0].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[0, 0].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')
axes[0, 1].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[0, 1].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')
axes[1, 0].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[1, 0].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')
axes[1, 1].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[1, 1].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')
axes[2, 0].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[2, 0].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')
axes[2, 1].set_xlabel('Time steps (8 s intervals)' ,fontsize = '30', color='black',weight='bold')
axes[2, 1].set_ylabel('Power (W)',fontsize = '30', color='black',weight='bold')

axes[0,0].set_title('Healthy',fontweight="bold", size=25) # Title
axes[0,1].set_title('Unhealthy with High Energy',fontweight="bold", size=25) # Title
axes[1,0].set_title('Unhealthy with Extended low energy Cycle',fontweight="bold", size=25) # Title
axes[1,1].set_title('Unhealthy with Repeated Cycles',fontweight="bold", size=25) # Title
axes[2,0].set_title('Unhealthy with Noise 1',fontweight="bold", size=25) # Title
axes[2,1].set_title('Unhealthy with Noise 2',fontweight="bold", size=25) # Title


# Show the plot
fig.suptitle("Examples of Healthy/Unhealthy (Anomalous) activations for appliance Dishwasher", fontsize=30)
fig.patch.set_linewidth(10)
plt.tight_layout(rect=[0, 0, 0.99, 0.99])
plt.show()
```
